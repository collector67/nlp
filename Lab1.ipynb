```python
import warnings
warnings.filterwarnings('ignore')

from deeppavlov import build_model, configs
model = build_model('rusentiment_bert', download=False)

from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="blanchefort/rubert-base-cased-sentiment"
)

texts = [
    "Отвратительй день выдался сегодня",
    "Ужасная погода сегодня",
    "Утро сегодня просто отличное",
    "Мне не очень нравится это",
    "Какой прекрасный день!",
    "Дорогой друг,с уважением благодарю тебя)"
]

results = classifier(texts)

sentiment_map = {
    'POSITIVE': 'позитивная',
    'NEGATIVE': 'негтивная',
    'NEUTRAL': 'нейтральная'
}

for text, result in zip(texts, results):
    sentiment_en = result['label']
    sentiment_ru = sentiment_map.get(sentiment_en, sentiment_en)
    confidence = result['score']

    print(f"Текст: {text}")
    print(f"Тональность: {sentiment_ru} (уверенность: {confidence:.2f})")
    print("-" * 50)
```
