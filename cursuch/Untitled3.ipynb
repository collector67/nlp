```python
import ollama

response = ollama.list()
print("Структура ответа:", type(response))   # для отладки

print("\nУстановленные модели:")
for model in response.models:          # ← .models, а не ['models']
    name = model.model                 # ← .model, а не ['name']
    param_size = model.details.parameter_size if hasattr(model, 'details') else 'N/A'
    print(f"• {name}  (параметров: {param_size})")
```
