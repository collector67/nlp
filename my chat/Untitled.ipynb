```python
import ollama

MODEL = "deepseek-r1:14b"

print("Чат-бот DeepSeek-R1 (14B) успешно запущен!")
print("Пиши 'выход', 'exit', 'quit' или 'q' для завершения.\n")

history = []

system_prompt = {
    "role": "system",
    "content": "Ты — полезный, дружелюбный и немного остроумный помощник. Отвечай всегда на русском языке."
}

while True:
    user_input = input("Ты: ")
    
    if user_input.lower() in ["выход", "exit", "quit", "q"]:
        print("До встречи!")
        break
    
    # Формируем историю + текущее сообщение
    messages = [system_prompt]
    
    for user_msg, assistant_msg in history:
        if user_msg:
            messages.append({"role": "user", "content": str(user_msg)})
        if assistant_msg:
            messages.append({"role": "assistant", "content": str(assistant_msg)})
    
    messages.append({"role": "user", "content": user_input})
    
    print("DeepSeek думает... (может занять 10–40 секунд)")
    
    response = ollama.chat(model=MODEL, messages=messages)
    answer = response['message']['content']
    
    print(f"DeepSeek: {answer}\n")
    
    # Сохраняем в историю
    history.append((user_input, answer))
```
